# １. Rを動かしてみよう
それでは、実際にRを動かしてみましょう。以下のテキストセルの下にコードセルを追加し、出力結果を予想しながら、以下のコマンドを実行してみてください。


```R
############################################################
# 以下のコマンドをコードセルに入力し、実行してみてください #
############################################################
1+1
(1+2)*3
7/2
7%%2
7%/%2
5^2
sin(pi/2)
exp(1)
sqrt(16)
```


Rは統計プログラミングに特化したコンピュータ言語ですので、様々な数学関数が標準で利用できるようになっています。上記コマンドでそのようすが確認できたと思います。




次に、特定の値や計算結果を変数に代入し、保持しておく方法を試してみましょう。値を保持しておくための変数のことをRでは「オブジェクト」と呼び、数値だけでなくベクトルや行列、関数などを入れておくことができます。一連の処理の中で結果を一時的に保存しておいたり、再利用するような場面で利用します。まず、変数xとyに10という数値を代入してみましょう。


```R
############################################################
# 以下のコマンドをコードセルに入力し、実行してみてください #
############################################################
x = 10
y = 10
x
```


オブジェクト名`x`を実行すると、その直前のコマンドで代入しておいた`10`が出力されたと思います。
オブジェクトに与えられた値は、上書きしないか、ランタイムがリセットされない限りR上に保持されます。




オブジェクト同士の計算も可能で、その結果をさらに別のオブジェクトに代入することもできます。いま`x`と`y`にはそれぞれ`10`が代入された状態になっていますが、これらを利用して、オブジェクトどうしの計算を行ってみましょう。


```R
############################################################
# 以下のコマンドをコードセルに入力し、実行してみてください #
############################################################
x + y
z = x * y
z
```


比較演算子を利用すると、数の大きさを比べることができます。結果はその比較評価式が正しいのか、あるいは間違っているのかが論理値（`TRUE` or `FALSE`）形式で返ってきます。比較演算子は関数処理の中の条件分岐等で利用される演算子です。詳細については関数の講義で再度取り上げたいと思います。


```R
############################################################
# 以下のコマンドをコードセルに入力し、実行してみてください #
############################################################
x <= y
x == y
x > y
```


値が並んだ配列データのことをベクトルと呼び、Rはベクトルを基本にしてあらゆるものが設計されています。これまで試してきたいくつかのRの実行例も、実は長さ1のベクトルとみなすことができます。

ベクトルの作り方にはいくつかの方法がありますが、一番簡単な方法は`c()`という関数を使った作り方です。以下の例では長さ3の数値ベクトルと、長さ2の文字列ベクトルを作っています。ダブルクオーテーション"で文字列を囲うことで、文字列データを生成することができます。


```R
############################################################
# 以下のコマンドをコードセルに入力し、実行してみてください #
############################################################
c(1, 2, 4)
c("hello", "thank you")
```


# ２. Rのヘルプ機能
本講座では今後さまざまなRの関数を利用していきますが、指定可能な引数の詳細など、関数の使い方のすべてを講義内で説明することは困難です。もし講義で取り上げた関数について疑問点等があれば、Rのヘルプ機能を利用してみてください。`?関数名`あるいは、`help(関数名)`のように入力すると、画面右側に関数の引数の説明、出力内容、コマンド実行例などのヘルプ情報が表示されますので参考にしてみてください。


```R
############################################################
# 以下のコマンドをコードセルに入力し、実行してみてください #
############################################################
?read.table
help(read.table)
```
